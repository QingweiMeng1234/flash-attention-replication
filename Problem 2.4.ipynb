```python
# Problem 2.4.3
import matplotlib.pyplot as plt

x = range(1,6)
bleu_score = [37.923525691353774, 38.70245035207466, 38.70326091637565, 38.514007540087356, 38.44263374237749]

# Create the plot
plt.figure(figsize=(10,6))
plt.plot(x, bleu_score, marker='o')

# Add titles and labels
plt.title('BLEU Score vs Beam Size')
plt.xlabel('Beam Size')

plt.ylabel('BLEU Score')

# Display the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
